```python
%%cython

from libc.math cimport sqrt

def sweep(float[:,:] A, float[:,:] Cost):
    cdef int i, j
    cdef float t0, t1, t2, C
    cdef int count = 0
    for i in xrange(1, A.shape[0]):
        for j in xrange(1, A.shape[1]):
            t1, t2 = A[i, j-1], A[i-1, j]
            C = Cost[i, j]
            if abs(t1-t2) > C:
                t0 = min(t1, t2) + C
            else:    
                t0 = 0.5*(t1 + t2 + sqrt(2*C**2 - (t1-t2)**2))
            if t0 < A[i, j]:
                A[i, j] = t0
                count += 1
    print count
    return count
```
