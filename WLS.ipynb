```python
import scipy.sparse as ssp
import scipy.sparse.linalg
figsize(10, 8)
```

```python
npz = load('disp.npz')

A = npz['disp']
W = npz['a'] * 0.01
h, w = A.shape
```

```python
idx = arange(A.size).reshape(*A.shape)
J = idx[:,:-1]

I, J, D = [], [], []
row = 0
for i, j in ndindex(A.shape):
    if j < w-1:# and not (j == 30 and i > 30):
        I.extend([row, row])
        J.extend([i*w+j, i*w+j+1])
        D.extend([-1, 1])
        row += 1
    if i < h-1:
        I.extend([row, row])
        J.extend([i*w+j, i*w+j+w])
        D.extend([-1, 1])
        row += 1
D = ssp.csr_matrix((D, (I, J)))
```

```python
M = ssp.spdiags(W.ravel()**2, 0, w*h, w*h) + D.T*D
X = ssp.linalg.spsolve(M, (A*W*W).ravel()).reshape(h, w)
```
