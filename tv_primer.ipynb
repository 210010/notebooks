Dual:

$$ F^*(p) = \sup_ {x \in X} \; \langle p, x \rangle - F(x)$$

$F = F^{**}$

$$ \min_x \; F(Kx) + G(x) $$

Form 1:

$$ \min_x \; \max_p  \; \langle Kx, p\rangle  + G(x) - F^*(p)$$

Form 2:

$$ \min_x \; \max_{p, r}  \; \langle Kx, p\rangle  + \langle x, r \rangle - F^\ast(p) - G^\ast(r) $$

Denoiseing
----------

* ROF

$$ \min_u \lambda \|\nabla u\| + \|u - f\|^2 $$

* TV-L1

$$ \min_u \lambda \|\nabla u\| + \|u - f\| $$

Binary segmentation
-------------------

$$ \min_u g \|\nabla u\| + c_b u + c_f (1 - u) $$
$$ \min_u g \|\nabla u\| + uf $$

Stereo
------

$$ \min_u \|\nabla u(\mathbf{x})\| + \rho(\mathbf{x}, u(\mathbf{x}))  $$

$$ \min_\phi \|\nabla_\mathbf{x} \phi(\mathbf{x}, \lambda)\| + 
\rho(\mathbf{x}, \lambda)\left| \frac{\partial}{\partial \lambda}\phi(\mathbf{x}, \lambda) \right|  $$
