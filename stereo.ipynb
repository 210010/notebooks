```python
def box(a, r):
    return cv2.boxFilter(a, cv2.CV_32F, (2*r+1, 2*r+1), normalize=False, borderType=cv2.BORDER_CONSTANT)

def guided_filter(I, p, r, eps):
    N = box(ones(I.shape[:2], np.float32), r)
    
    mean_I = box(I, r) / N
    mean_p = box(p, r) / N
    mean_Ip = box(I*p, r) / N
    cov_Ip = mean_Ip - mean_I * mean_p
    
    mean_II = box(I*I, r) / N
    var_I = mean_II - mean_I * mean_I
    
    a = cov_Ip / (var_I + eps)
    b = mean_p - a * mean_I
    
    mean_a = box(a, r) / N
    mean_b = box(b, r) / N
    
    q = mean_a * I + mean_b
    return q
```

```python
def make_stereo(dmin, dmax, P1=40, P2=400): # 
    window_size = 3
    dgran = 16
    dmin = int(dmin // dgran * dgran)
    dmax = int((dmax+dgran-1) // dgran * dgran)
    stereo = cv2.StereoSGBM(minDisparity = dmin, numDisparities = dmax-dmin, SADWindowSize = window_size,
        uniquenessRatio = 15,
        speckleWindowSize = 100,
        speckleRange = 32,
        disp12MaxDiff = 1,
        P1 = P1*window_size**2,
        P2 = P2*window_size**2,
        preFilterCap = 100,
        fullDP = True)
    return stereo
```

```python
st = make_stereo(-dmax, -dmin, P1=40, P2=400)
Dsg = st.compute(np.uint8(img_r*255), np.uint8(img_l*255))

vis = -np.float32(Dsg)
vis = (vis - vis.min()) / 500.0
vis[vis == vis.max()] = 0
show_array(vis)
```
