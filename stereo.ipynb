```python
import pyximport; pyximport.install()

import scipy.ndimage as nd
import cv2
from nbutils import showarray
from guided import guided_filter_rgb

figsize(10, 8)
```

```python
img_r = np.float32(cv2.imread('data/03_right.bmp')) / 255.0
img_l = np.float32(cv2.imread('data/03_left.bmp')) / 255.0
img_l = img_l[256:, 256:]
img_r = img_r[256:, 256:]
```

```python
h, w = img_r.shape[:2]
dmin, dmax, dstep = -10, 90, 1.0
ds = arange(dmin, dmax, dstep)
dnum = len(ds)

vis = np.hstack([img_r, img_l])
vis = cv2.cvtColor(vis, cv2.COLOR_BGR2RGB)
showarray(vis)
```

```python
color_thresh = 0.1

R = np.zeros((dnum, h, w), np.float32)

for i, d in enumerate(ds):
    print d,
    M = np.column_stack([eye(2), [-d, 0]])
    img_ld = cv2.warpAffine(img_l, M, (w, h), borderMode=cv2.BORDER_REPLICATE)
    Ri = cv2.absdiff(img_ld, img_r)
    Ri = cv2.cvtColor(Ri, cv2.COLOR_BGR2GRAY)
    Ri = minimum(Ri, color_thresh)
    Ri = guided_filter_rgb(img_r, Ri, 15, 0.1**2)
    R[i] = Ri
    
D = R.argmin(0)
   
```

```python
showarray(R[:,300]/R.max())
showarray(1.3*D/dnum)
```

```python
i, j = indices(D.shape)
r0 = R[np.maximum(D-1, 0), i, j]
r1 = R[D, i, j]
r2 = R[np.minimum(D+1, dnum-1), i, j]
dr = -0.5 * (r2-r0) / (r0 - 2*r1 + r2)

Ds = D + dr
showarray(1.3*Ds/dnum)
```

```python
def make_stereo(dmin, dmax, P1=40, P2=400): # 
    window_size = 3
    dgran = 16
    dmin = int(dmin // dgran * dgran)
    dmax = int((dmax+dgran-1) // dgran * dgran)
    stereo = cv2.StereoSGBM(minDisparity = dmin, numDisparities = dmax-dmin, SADWindowSize = window_size,
        uniquenessRatio = 15,
        speckleWindowSize = 100,
        speckleRange = 32,
        disp12MaxDiff = 1,
        P1 = P1*window_size**2,
        P2 = P2*window_size**2,
        preFilterCap = 100,
        fullDP = True)
    return stereo
```

```python
st = make_stereo(-dmax, -dmin, P1=40, P2=400)
Dsg = st.compute(np.uint8(img_r*255), np.uint8(img_l*255))

vis = -np.float32(Dsg)
vis = (vis - vis.min()) / 500.0
vis[vis == vis.max()] = 0
showarray(vis)
```
