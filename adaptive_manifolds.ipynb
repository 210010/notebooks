```python
def blur(a):
    return cv2.GaussianBlur(a, (25, 25), -1)

def am_split(img, mask):
    img_blur = blur(img*mask)
    mask_blur = blur(mask)
    m = cv2.divide(img_blur, mask_blur)
    lo_mask = float32(img <= m)
    hi_mask = 1.0 - lo_mask
    return m, lo_mask, hi_mask
    
```

```python
def am_build(img, mask, depth):
    m, lo_mask, hi_mask = am_split(img, mask)
    if depth == 1:
        return [m]
    lo = am_build(img, lo_mask, depth-1)
    hi = am_build(img, hi_mask, depth-1)
    return lo + [m] + hi
```

```python
mf = am_build(img, ones_like(img), 3)
print len(mf)
```

```python
sigma_c = 0.1

def am_project(img, m):
    w = exp(-(img - m)**2/(sigma_c**2))
    bw, bwimg = blur(w), blur(w*img)
    img = bwimg / bw
    img = bwimg*w
    return w, img

w, c = sum([am_project(img, m) for m in mf], 0)
res = c/w
```
