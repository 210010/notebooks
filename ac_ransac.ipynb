```python
def sample_line(p1, p2, n, noise=0.0):
    p1 = np.float32(p1)
    t = np.random.rand(n,1)
    return p1 + (p2-p1)*t + np.random.normal(size=(n, 2))*noise
```

```python
N = 200
inlier_n = int(0.7*N)
inliers = sample_line((1, 3), (6, 1), inlier_n, 0.05)
outliers = rand(N-inlier_n, 2)*(7, 4)
points = np.vstack([inliers, outliers])
shuffle(points)
```

```python
def fitline(p):
    n = len(p)
    A = column_stack([p, np.ones(n)])
    U, S, V = svd(A.T.dot(A))
    return V[2]
    
def residuals(C, p):
    a, b, c = C
    x, y = p.T
    return a*x + b*y + c
```

```python
def draw_line(C, **kw):
    a, b, c = C
    plot((0, -c/a), (-c/b, 0), **kw)
```

```python
x, y = points.T
plot(x, y, '.')
draw_line(fitline(inliers), color='green')
draw_line(fitline(points), color='red')
```

```python
r = residuals(fitline(inliers), points)
r = sort(abs(r))
plot(r)
```

```python
r = residuals(fitline(points), points)**2
r = sort(abs(r))
plot(r)
```

```python
def estimate_threshold(r, inliner_prob_f):
    r = sort(abs(r))
    n = len(r)
    k = arange(1, n+1)
    p = inliner_prob_f(r)
    score = logCnk(n, k)+log(p)*k#+log(1.0-p)*(n-k)
    return score

plot(estimate_threshold(residuals(fitline(inliers), points), lambda r:r))
plot(estimate_threshold(residuals(fitline(points), points), lambda r:r))
plot(estimate_threshold(residuals(fitline(outliers), points), lambda r:r))
```
