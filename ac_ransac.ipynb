```python
def sample_line(p1, p2, n, noise=0.0):
    p1 = np.float32(p1)
    t = np.random.rand(n,1)
    err = np.random.normal(size=(n, 2))
    #err = (2*rand(n, 2)-1)
    return p1 + (p2-p1)*t + err*noise
```

```python
N = 1000
inlier_n = int(0.1*N)
inliers = sample_line((1, 3), (6, 1), inlier_n, 0.05)
outliers = rand(N-inlier_n, 2)*(7, 4)
points = np.vstack([inliers, outliers])
shuffle(points)
```

```python
def fitline(p):
    n = len(p)
    A = column_stack([p, np.ones(n)])
    U, S, V = svd(A.T.dot(A))
    return V[2]
    
def residuals(C, p):
    a, b, c = C
    x, y = p.T
    return a*x + b*y + c
```

```python
def draw_line(C, **kw):
    a, b, c = C
    plot((0, -c/a), (-c/b, 0), **kw)
```

```python
n = 10000
logF = np.zeros(n)
logF[1:] = log(arange(1, n)).cumsum()

def logCnk(n, k):
    return logF[n]-logF[k]-logF[n-k]
    
```

```python
def estimate_threshold(r):
    r = sort(abs(r))
    n = len(r)
    k = arange(1, n+1)
    p = r*pi/(4.0*7.0)
    scores = log(r)*k + logCnk(n, k)# + logCnk(k, 4)
    return scores

plot(estimate_threshold(residuals(fitline(inliers), points)))
plot(estimate_threshold(residuals(fitline(points), points)))
plot(estimate_threshold(residuals(fitline(outliers), points)))
```

```python
def residuals_s(C, p):
    return sort(abs(residuals(C, p)))
plot(residuals_s(fitline(inliers), points))
plot(residuals_s(fitline(points), points))
plot(residuals_s(fitline(outliers), points))
```
