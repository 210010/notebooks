```python
datadir = 'data/roofs/'

img_l = cv2.imread(datadir+'img_l.bmp')
img_r = cv2.imread(datadir+'img_r.bmp')
x0, y0, w, h, d1, d2 = loadtxt(datadir+'roi.txt', int32)
meta = pickle.load(open(datadir+'Q.dat'))
Q = meta['Q']

roi = (x0, y0), (x0+w, y0+h)
drange = d1, d2
```

```python
subimg_l = img_l[y0:,x0:][:h,:w]
imshow(subimg_l[...,::-1])
```

```python
def cost_rgb(img_l, img_r, roi_l, drange):
    dmin, dmax = drange
    dnum = dmax - dmin
    (x1, y1), (x2, y2) = roi_l
    w, h = x2-x1, y2-y1
    R = np.zeros((h, w, dnum), np.float32)
    sub_l = img_l[y1:, x1:][:h, :w]
    for i in xrange(dnum):
        print i,
        sub_r = img_r[y1:, x1-i-dmin:][:h, :w]
        diff = cv2.absdiff(sub_l, sub_r)
        for ch in xrange(diff.shape[-1]):
            R[...,i] += diff[..., ch]
    return R
```

```python
R = cost_rgb(img_l, img_r, roi, drange)
#R = minimum(R, 32*3)
#R[R < 30] = 1
imshow(R[300, 100:250].T)
```

```python
from sgm import SGM, calc_cost

#Cost = SGM(R, 20, 200)    
Cost = zeros_like(R)
calc_cost(R, Cost, 100, 1000, 0, 1)    
     
imshow(Cost[300, 100:250].T)
```

```python
def subpix_disparity(Cost):
    D = Cost.argmin(-1)
    i, j = indices(D.shape)
    r0 = Cost[i, j, np.maximum(D-1, 0)]
    r1 = Cost[i, j, D]
    r2 = Cost[i, j, np.minimum(D+1, Cost.shape[2]-1)]
    eps = 1e-7
    return np.float32( D - 0.5 * (r2-r0) / (r0 - 2*r1 + r2 + eps) )

def save_cloud(fn, D):
    p3d = cv2.reprojectImageTo3D(D, Q)
    ply.write_ply_bin(fn, p3d, subimg_l[...,::-1])

Disp = subpix_disparity(Cost)
save_cloud('cloud.ply', Disp)
imshow(Disp)
```
