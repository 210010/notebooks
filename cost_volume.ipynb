```python
%load_ext cythonmagic
import pyximport; pyximport.install()
import cv2
from nbutils import showarray
import cPickle as pickle
import ply
from sgm import *
figsize(10, 8)
```

```python
datadir = 'data/plant/'

img_l = cv2.imread(datadir+'img_l.bmp')
img_r = cv2.imread(datadir+'img_r.bmp')
x0, y0, w, h, dmin, dmax = loadtxt(datadir+'roi.txt', int32)
meta = pickle.load(open(datadir+'Q.dat'))
Q = meta['Q']

roi = (x0, y0), (x0+w, y0+h)
dnum = dmax-dmin

img_l = img_l[y0:,x0:][:h,:w]
img_r = img_r[y0:,x0-dmax:][:h,:w+dnum]
```

```python
vis = hstack([img_r, img_l])
showarray(vis[...,::-1])
```

```python
def cost_rgb(img_l, img_r, dnum):
    h, w = img_l.shape[:2]
    R = np.zeros((h, w, dnum), np.float32)
    for i in xrange(dnum):
        print i,
        sub_r = img_r[:, dnum-i-1:][:, :w]
        diff = cv2.absdiff(img_l, sub_r)
        for ch in xrange(diff.shape[-1]):
            R[...,i] += diff[..., ch]
    return R

def cost_canny(img_l, img_r, dnum):
    def preprocess(a):
        return cv2.Canny(a, 64, 100)
    img_l, img_r = map(preprocess, [img_l, img_r])
    
    h, w = img_l.shape[:2]
    R = np.zeros((h, w, dnum), np.float32)
    for i in xrange(dnum):
        print i,
        sub_r = img_r[:, dnum-i-1:][:, :w]
        R[...,i] = cv2.absdiff(img_l, sub_r)
    return R
```

```python
R = cost_rgb(img_l, img_r, dnum)
#R = cost_rgb(img_l, img_r, dnum)
R /= 2.0*R.mean()
#R = minimum(R, 1.0)
#save('Cost.npy', R)
imshow(R[300, 100:250].T)
```

```python
Cost = SGM(1.0, 10)(R)#, paths=[(0, 1), (0, -1)])
imshow(Cost[300, 100:250].T)
```

```python
def subpix_disparity(Cost):
    D = Cost.argmin(-1)
    i, j = indices(D.shape)
    r0 = Cost[i, j, np.maximum(D-1, 0)]
    r1 = Cost[i, j, D]
    r2 = Cost[i, j, np.minimum(D+1, Cost.shape[2]-1)]
    eps = 1e-7
    a = np.float32( 0.5 * (r0+r2-2*r1) )
    D = np.float32( D - 0.5 * (r2-r0) / (r0 - 2*r1 + r2 + eps) )
    return D, a

def save_cloud(fn, D):
    p3d = cv2.reprojectImageTo3D(D, Q)
    ply.write_ply_bin(fn, p3d, img_l[...,::-1])

Disp, a = subpix_disparity(Cost)
savez('disp.npz', disp=Disp, a=a)
save_cloud('cloud.ply', Disp)
save_cloud('cloud2.ply', float32(load('disp2.npy')))
imshow(Disp)
```
